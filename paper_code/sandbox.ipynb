```python
import random
import torch
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.transforms.functional import to_pil_image
```

```python
index = random.randint(3, 63)

fig, axes = plt.subplots(ncols=3, nrows=1)

path_preloaded = f'preloaded_adv_images/adv_{index}.pt'
path_original = f'./init_images/{index}.png'
path_mask = f'./init_images/{index}_maskprocessed_mask.png'
preloaded_adv_image = to_pil_image(torch.load(path_preloaded, map_location='cpu')[0])
labels = [path_preloaded, path_original, path_mask]
images : [Image] = [preloaded_adv_image, Image.open(path_original), Image.open(path_mask)]
for i, (ax, img) in enumerate(zip(axes.flatten(), images)):
    ax.imshow(img)
    ax.tick_params(axis='both', which='both', length=0, labelsize=0)
    ax.set_xlabel(labels[i])

fig.set_size_inches(15, 6)
plt.set_cmap('gray')
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
